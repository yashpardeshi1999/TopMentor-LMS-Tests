```python
# Filter the first 10 restaurants
first_10 = df.head(10)

# Find the most expensive restaurants
most_expensive = first_10[first_10['Average Cost for two'] == first_10['Average Cost for two'].max()]
print(most_expensive[['City', 'Average Cost for two']])
```

```python
# Find the cheapest restaurants
cheapest = first_10[first_10['Average Cost for two'] == first_10['Average Cost for two'].min()]
print(cheapest[['City', 'Average Cost for two']])
```

```python
# Find the highest rated restaurants
highest_rated = first_10[first_10['Aggregate rating'] == first_10['Aggregate rating'].max()]
print(highest_rated[['City', 'Aggregate rating']])
```

```python
# Find restaurants without table booking
no_table_booking = first_10[first_10['Has Table booking'] == 'No']
print(no_table_booking[['Restaurant Name', 'Has Table booking']])
```

```python
# Filter the rows for Indian restaurants between index 90 and 1405 (excluding)
indian_restaurants = df.iloc[90:1405]

# Filter only Indian restaurants (assuming the "Cuisines" column contains the type of cuisine)
indian_restaurants = indian_restaurants[indian_restaurants['Cuisines'].str.contains('Indian', case=False, na=False)]

# 1. Indian restaurants with less than 4/5 rating
low_rated_restaurants = indian_restaurants[indian_restaurants['Aggregate rating'] < 4]['Restaurant Name']
print("Less than 4 Rated Restaurants:", low_rated_restaurants.values)
```

```python
# 2. Most expensive Indian restaurant
most_expensive_indian = indian_restaurants[indian_restaurants['Average Cost for two'] == indian_restaurants['Average Cost for two'].max()]['Restaurant Name']
print("Most Expensive Restaurant:", most_expensive_indian.values)
```

```python
# 3. Cheapest Indian restaurant
cheapest_indian = indian_restaurants[indian_restaurants['Average Cost for two'] == indian_restaurants['Average Cost for two'].min()]['Restaurant Name']
print("Cheapest Restaurant:", cheapest_indian.values)
```
