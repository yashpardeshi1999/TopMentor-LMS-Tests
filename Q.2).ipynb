#  Top Two Job Profiles

```python
# Count the occurrences of each job type in the first 20 rows
top_jobs = data['job'].head(20).value_counts().nlargest(2)
print("Top two job profiles:\n", top_jobs)
```

# Education Level and Loan

```python
# Filter the first 20 people and group by education level and loan status
education_loan = data[['education', 'loan']].head(20)

# Count the number of people with secondary and tertiary education who opted for a loan
loan_by_education = education_loan[education_loan['loan'] == 'yes']['education'].value_counts()
print("Number of people with secondary or tertiary education who opted for a loan:\n", loan_by_education)
```
